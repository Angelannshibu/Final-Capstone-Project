DATA PREPROCESSING

```python
df.isnull().sum()
#Checks for null values and found zero null values
```

```python
df = df.drop_duplicates()
#Removes duplicate rows
```

```python
df['Final_Grade'] = (df['G1'] + df['G2'] + df['G3']) / 3
#Calculates average grade
```

```python
df['Result'] = df['G3'].apply(lambda x: 'Pass' if x >= 10 else 'Fail')

#Classifies students:
#Pass (G3 ≥ 10) and Fail (G3 < 10)
```

Exploratory Data Analysis (EDA) and Visualizations


EDA is performed to identify patterns, trends, and relationships using statistical analysis and visual exploration.


## 1. UNIVARIATE ANALYSIS
#This section analyzes individual variables to understand their distribution, central tendency, and spread.



```python
#Grade Distribution 
#Histogram shows how final grades are distributed among students.

print(df['G3'].describe())

plt.hist(df['G3'], bins=10) 
plt.title("Distribution of Final Grades")
plt.xlabel("Final Grade")
plt.ylabel("Frequency") 
plt.show()
```

#Insight:The histogram and summary statistics show how final grades (G3) are distributed among students. Most students score within the middle grade range, indicating that average performance is common across the dataset. Extremely low and extremely high scores appear less frequently.This pattern suggests a balanced distribution of academic performance,

Absence Distribution

```python
#Shows attendance behavior using absences data.

plt.hist(df['absences'], bins=10)
plt.title("Absence Distribution")
plt.xlabel("Absences")
plt.ylabel("Frequency")
plt.show()
```

#Insight:
This histogram shows how frequently students are absent.
Most students have low absence counts, indicating good attendance behavior. 
However, a small group has very high absences, which may negatively affect their academic performance and require attention.

Gender vs Performance

```python
#Average grade comparison between male and female students.

df.groupby('sex')['G3'].mean()
```

#Insight:
Shows whether performance differs between male and female students.The average final grade (G3) for female students is 12.25, while for male students it is 11.41. This indicates that female students, on average, perform slightly better than male students in this dataset.This analysis helps understand performance trends across genders

Study Time vs Grade

```python
#Shows relationship between study time and performance.

df.groupby('studytime')['G3'].mean()
```

#Insight:Students who study more hours tend to have higher grades.The grouped analysis shows that students with higher study time levels generally achieve better average final grades (G3). As study time increases, the average performance also shows an upward trend.

Failures vs Grade

```python
#Students with more failures tend to have lower grades.

df.groupby('failures')['G3'].mean()
```

#Insight:The grouped analysis shows that as the number of previous failures increases, the average final grade (G3) decreases significantly. Students with no past failures tend to have higher average grades, while those with one or more failures show noticeably lower performance.This indicates that academic history strongly influences current performance.

## 2. BIVARIENT ANALYSIS

```python
#Absences vs Grade
#Higher absences often correlate with lower grades.

plt.scatter(df['absences'], df['G3'])
plt.title("Absences vs Final Grade")
plt.xlabel("Absences")
plt.ylabel("Final Grade")
plt.show()
```

#Insight:The scatter plot shows a negative relationship between the number of absences and the final grade (G3). Students with fewer absences tend to achieve higher grades, while those with frequent absences often have lower academic performance.
Although the relationship is not perfectly linear, the overall trend suggests that regular attendance plays an important role in maintaining consistent academic performance.

```python
#Study Time vs Grade

sns.boxplot(x='studytime', y='G3', data=df)
plt.title("Study Time vs Final Grade")
plt.show()
```

#Insight:The box plot shows how final grades (G3) vary across different study time levels. Students in higher study time categories generally have higher median grades compared to those who study less. This suggests that increased study time contributes to better academic performance.

## 3. MULTIVARIATE ANALYSIS

```python
corr = df[['G1','G2','G3','absences','studytime','failures']].corr()

sns.heatmap(corr, annot=True)
plt.title("Correlation Matrix")
plt.show()
```

#Insight:The correlation heatmap shows the relationships between academic variables such as G1, G2, G3, absences, study time, and failures. There is a strong positive correlation between G1, G2, and G3, indicating that internal assessment grades are strong predictors of final performance.Failures show a negative correlation with final grades, meaning students with more past failures tend to perform worse. Absences also show a slight negative relationship with grades, while study time has a mild positive correlation. Overall, the heatmap highlights that internal grades are the most influential factors in predicting final academic performance.

```python
pivot = pd.pivot_table(
    df,
    values='G3',
    index='studytime',
    columns='failures',
    aggfunc='mean'
)

print(pivot)
```

#Insight:The pivot table shows the average final grade (G3) based on different combinations of study time and number of past failures. Students with higher study time and zero failures consistently achieve the highest average grades, indicating that consistent effort combined with a strong academic history leads to better performance.

```python
#Students with high absences and failures

risk_students = df[(df['absences'] > 10) & (df['failures'] > 0)]
print(len(risk_students))
```

#Insight:This analysis identifies students who have both high absences (more than 10) and at least one previous failure. These students are more likely to experience academic difficulties because they combine two major risk factors: poor attendance and weak academic history.Monitoring these students and providing targeted support,can help improve their performance and reduce the likelihood of future failures.

## 4. GROUPBY

```python
#Performance by Gender
df.groupby('sex')['G3'].mean()
```

#Insight:The analysis compares the average final grade (G3) between male and female students. The results show that female students have a slightly higher average final grade than male students, indicating marginally better academic performance in this dataset.

```python
#Average Grade by Study Time
df.groupby('studytime')['G3'].mean()
```

#Insight:The grouped results indicate a clear upward trend in average final grades (G3) as study time increases. Students in lower study time categories tend to have lower average scores, while those in higher study time categories achieve better academic outcomes.This demonstrates that consistent and increased study effort positively influences performance.

```python
#Failures vs Final Grade
df.groupby('failures')['G3'].mean()
```

#Insight:The grouped analysis shows that the average final grade (G3) decreases as the number of previous failures increases. Students with no prior failures tend to achieve the highest average grades, while those with one or more failures show progressively lower performance.This suggests that previous academic difficulties continue to influence current performance.

```python
#VISUALIZATION
#Study Time Distribution
#Shows frequency of students in each study time category.

df['studytime'].value_counts()

sns.countplot(x='studytime', data=df)
plt.title("Study Time Levels")
plt.show()
```

#Insight:The count plot shows the number of students in each study time category. Most students fall into the moderate study time levels, indicating that the majority maintain a balanced study routine. Only a small number of students belong to the highest study time category.This distribution suggests that extreme study patterns are less common among students.

```python
#Final Grade Distribution (Histogram)

sns.countplot(x='G3', data=df)
plt.title("Pass vs Fail Distribution")
plt.xlabel("G3")
plt.ylabel("Number of Students")
plt.show()
```

#Insight:The histogram shows how student final grades (G3) are distributed among students.Most students fall in the medium score range, while very few students achieve extremely high or very low grades.This pattern suggests a balanced grade distribution, where extreme outcomes are less common

```python
import matplotlib.pyplot as plt

# Create Pass/Fail column (if not already created)
df['Result'] = df['G3'].apply(lambda x: 'Pass' if x >= 10 else 'Fail')

result_counts = df['Result'].value_counts()

plt.figure(figsize=(5,5))
plt.pie(result_counts.values,
        labels=result_counts.index,
        autopct='%1.1f%%',
        startangle=90)

# Draw center circle to make it donut
centre_circle = plt.Circle((0,0), 0.60, fc='white')
plt.gca().add_artist(centre_circle)

plt.title("Pass vs Fail Distribution")
plt.show()
```

#Insight:The donut chart shows the proportion of students who passed and failed based on their final grade (G3). The visualization indicates that most students fall into the pass category, showing that overall academic performance is satisfactory.

```python
#Attendance Distribution

plt.hist(df['absences'], bins=10)
plt.title("Attendance (Absences) Distribution")
plt.xlabel("Number of Absences")
plt.ylabel("Frequency")
plt.show()
```

#Insight:The histogram shows that the majority of students have a low number of absences, indicating generally good attendance behavior across the dataset. However, there is a small group of students with significantly higher absence counts which can lead to academic risk.

```python
#G1 vs G2 Grade Comparison (Box Plot)

sns.boxplot(x='G1', y='G2', data=df)
plt.title("First grade vs Second grade")
plt.xlabel("G1")
plt.ylabel("G2")
plt.show()
```

#Insight:The box plot illustrates the relationship between the first period grade (G1) and the second period grade (G2). Students with higher G1 scores generally show higher G2 scores as well, indicating consistency in academic performance across internal assessments.

```python
#Study Time vs Final Grade (Scatter Plot)

sns.scatterplot(x='studytime', y='G3', hue='G2', data=df)
plt.title("Study Time vs Final Grade")
plt.xlabel("Study Time")
plt.ylabel("G3")
plt.show()
```

#Insight:The scatter plot illustrates the relationship between study time and the final grade (G3). Students who spend more time studying generally achieve higher final grades, showing a positive connection between study effort and academic performance.The color variation (G2) also indicates that strong internal performance often leads to better final results.


```python
#Internet Access vs Final Grade (Stacked Bar Chart)

internet_result = pd.crosstab(df['internet'], df['G3'])
internet_result.plot(kind='bar', stacked=True)

plt.title("Internet Access vs Result")
plt.xlabel("Internet Access")
plt.ylabel("Number of Students")
plt.show()
```

#Insight:The stacked bar chart compares the distribution of final grades (G3) between students who have internet access and those who do not. The visualization shows that students with internet access tend to achieve better final grades, suggesting that access to online resources may support academic performance.


```python
#Average Grade Progression (Line Plot)

avg_grades = df[['G1','G2','G3']].mean()

plt.plot(avg_grades.index, avg_grades.values, marker='o')
plt.title("Average Grade Progression")
plt.xlabel("Exam")
plt.ylabel("Average Score")
plt.show()
```

#Insight:The line plot shows the average scores of students across the three assessments — G1, G2, and G3. The values remain relatively close to each other, indicating that overall student performance is stable throughout the academic period.

```python
#Correlation Between Academic Factors

plt.figure(figsize=(10,6))
sns.heatmap(df[['G1','G2','G3','studytime','absences']].corr(),
            annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

#Insight:The correlation heatmap shows the strength of relationships between academic variables such as G1, G2, G3, study time, and absences. There is a strong positive correlation between G1, G2, and G3, indicating that students who perform well in internal assessments are very likely to achieve good final grades. This confirms that internal evaluation scores are strong indicators of overall academic performance.

```python
#Study Time Distribution

sns.countplot(x='studytime', data=df)
plt.title("Study Time Distribution")
plt.xlabel("Study Time Level")
plt.ylabel("Count")
plt.show()
```

#Insight:The count plot shows how students are distributed across different study time levels. Most students fall into the moderate study time categories, indicating that the majority maintain regular and balanced study habits. Only a small number of students are in the highest study time category.
